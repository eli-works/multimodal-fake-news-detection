```python
算法2-1 融合与分类流程

输入:
  text_seq      # 文本序列特征
  text_mask     # 文本有效位掩码
  image_vec     # 图像全局特征

步骤:
1. text_vec = MaskedMeanPool(text_seq, text_mask)
2. (text_share, image_share) = SharedBridge(text_vec, image_vec)
   # 目的：先校准图文表示，再进入交互
3. text_tokens = text_seq + text_share
4. (text_fused, image_fused) = BiCrossFusion(text_tokens, text_mask, image_share)
   # 目的：双向补充跨模态信息
5. fused = Concat(text_fused, image_fused)
6. logits = Classifier(fused)
7. probs = Softmax(logits)

输出:
  真假新闻标签与置信度
```
